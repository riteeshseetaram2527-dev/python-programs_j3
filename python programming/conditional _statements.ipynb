````python
"""

write a program to find the grades based on marks
1.if marks in between 80-100 - grade is A
2.if marks in between 60-79 - grade is B
3.if marks in between 40-59 - grade is c
4.if marks in between 25-39 - grade is D
5.if marks in between 0-24 - grade is fail
6.if marks other than above output is -please enter correct marks 
"""
marks =int(input("please enter student marks:"))
if (marks>=80 and marks<=100):
    print(marks,"grade is A")
elif(marks>=60 and marks<=79):
    print(marks,"grade is B")
elif(marks>=40 and marks<=59):
    print(marks,"grade is C")
elif(marks>=20 and marks<=39):
    print(marks,"grade is d")
elif(marks>=0 and marks<=19):
    print( marks,"grade is fail")
else:
    print("please enter correct marks")
````

````python
"""
write a program to find the even number with number of digits
1. if number is between 1 to 9 then output is - 1 digit even number
2. if number is between 10 to 99 then output is - 2 digit even number
3. if number is between 100 to 999 then output is - 3 digit even number
4. if number is between 1000 to 9999 then output is - 4 digit even number
5. if other than above range then output is - weird even number
"""

num = int(input("Enter a number: "))

if num % 2 == 0:
    if 1 <= num <= 9:
        print("1 digit even number")
    elif 10 <= num <= 99:
        print("2 digit even number")
    elif 100 <= num <= 999:
        print("3 digit even number")
    elif 1000 <= num <= 9999:
        print("4 digit even number")
    else:
        print("weird even number")
else:
    print("The number is not even.")
    
````
